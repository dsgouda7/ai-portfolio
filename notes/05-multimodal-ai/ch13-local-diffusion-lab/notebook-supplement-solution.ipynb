```python
import shutil\n
import subprocess\n
\n
def _gpu_available() -> tuple[bool, str]:\n
    # Prefer PyTorch when available.\n
    try:\n
        import torch  # type: ignore\n
        if torch.cuda.is_available():\n
            device_name = torch.cuda.get_device_name(0)\n
            return True, f'torch.cuda available ({device_name})'\n
        return False, 'torch installed, but cuda.is_available() is False'\n
    except Exception as ex:\n
        torch_reason = f'torch check unavailable: {ex}'\n
\n
    # Fallback to NVIDIA driver visibility.\n
    nvidia_smi = shutil.which('nvidia-smi')\n
    if not nvidia_smi:\n
        return False, f'{torch_reason}; nvidia-smi not found on PATH'\n
\n
    try:\n
        result = subprocess.run([nvidia_smi, '-L'], check=False, capture_output=True, text=True)\n
        detected = result.returncode == 0 and bool(result.stdout.strip())\n
        if detected:\n
            return True, 'GPU detected by nvidia-smi'\n
        stderr_text = result.stderr.strip() or 'no stderr output'\n
        return False, f'nvidia-smi did not report GPUs: {stderr_text}'\n
    except Exception as ex:\n
        return False, f'nvidia-smi check failed: {ex}'\n
\n
ok, reason = _gpu_available()\n
if not ok:\n
    raise SystemExit(\n
        'GPU required for this _Supplement notebook. '\n
        f'No compatible GPU was detected ({reason}).'\n
    )\n
\n
print(f'GPU check passed: {reason}')\n
```
