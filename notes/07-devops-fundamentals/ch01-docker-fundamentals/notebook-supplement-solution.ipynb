## Cell 1: Azure Credentials and Configuration

**Goal:** Configure Azure credentials and set resource names.

**Security:**
- Credentials are **stripped by pre-push hook** (see `scripts/install-hooks.ps1`)
- Never commit real subscription IDs or secrets
- Use empty strings as placeholders

**What we're configuring:**
- Azure subscription ID
- Resource group name
- Container registry name (must be globally unique)
- Region

````python
import os
import subprocess
import json

# Azure configuration (STRIPPED BY PRE-PUSH HOOK)
AZURE_SUBSCRIPTION_ID = ""  # Leave empty — get from: az account show --query id -o tsv
AZURE_RESOURCE_GROUP = "rg-devops-docker-demo"
AZURE_CONTAINER_REGISTRY = ""  # Must be globally unique, lowercase, alphanumeric only
AZURE_LOCATION = "eastus"

# Container configuration
IMAGE_NAME = "flask-app"
IMAGE_TAG = "v1"

print("Azure Configuration:")
print(f"  Subscription ID: {AZURE_SUBSCRIPTION_ID or '(not set)'}")
print(f"  Resource Group: {AZURE_RESOURCE_GROUP}")
print(f"  Registry Name: {AZURE_CONTAINER_REGISTRY or '(not set)'}")
print(f"  Location: {AZURE_LOCATION}")
print(f"  Image: {IMAGE_NAME}:{IMAGE_TAG}")

# Verify Azure CLI is logged in
try:
    result = subprocess.run(
        ["az", "account", "show"],
        capture_output=True,
        text=True,
        check=True
    )
    account = json.loads(result.stdout)
    print(f"\n✅ Logged in as: {account['user']['name']}")
    print(f"   Active subscription: {account['name']}")
    
    # Auto-fill subscription ID if not set
    if not AZURE_SUBSCRIPTION_ID:
        AZURE_SUBSCRIPTION_ID = account['id']
        print(f"   Auto-detected subscription ID: {AZURE_SUBSCRIPTION_ID[:8]}...")
        
except subprocess.CalledProcessError:
    print("\n❌ Not logged in to Azure. Run: az login")
    raise

# Verify registry name is set
if not AZURE_CONTAINER_REGISTRY:
    print("\n⚠️  AZURE_CONTAINER_REGISTRY not set.")
    print("   Choose a globally unique name (lowercase, alphanumeric):")
    print("   Example: acr" + os.urandom(4).hex())
    raise ValueError("AZURE_CONTAINER_REGISTRY must be set")

print("\n✅ Configuration valid. Ready to create Azure resources.")
````

## Cell 2: Create Azure Container Registry (ACR)

**Goal:** Create Azure Container Registry to host Docker images.

**What we're doing:**
1. Create resource group
2. Create Container Registry (Basic SKU — free tier for 12 months)
3. Enable admin credentials (for Docker login)
4. Retrieve registry credentials

**ACR SKUs:**
- **Basic** — 10 GB storage, 10 webhooks, good for dev/test ($0.17/day)
- **Standard** — 100 GB storage, geo-replication
- **Premium** — 500 GB storage, content trust, private endpoints

````python
%%bash -s "$AZURE_SUBSCRIPTION_ID" "$AZURE_RESOURCE_GROUP" "$AZURE_CONTAINER_REGISTRY" "$AZURE_LOCATION"

SUBSCRIPTION_ID=$1
RESOURCE_GROUP=$2
REGISTRY_NAME=$3
LOCATION=$4

# Set active subscription
az account set --subscription $SUBSCRIPTION_ID

# Create resource group
echo "Creating resource group: $RESOURCE_GROUP"
az group create \
  --name $RESOURCE_GROUP \
  --location $LOCATION

# Create Container Registry
echo ""
echo "Creating Container Registry: $REGISTRY_NAME"
az acr create \
  --resource-group $RESOURCE_GROUP \
  --name $REGISTRY_NAME \
  --sku Basic \
  --admin-enabled true

# Get registry login server
echo ""
echo "Registry details:"
LOGIN_SERVER=$(az acr show \
  --name $REGISTRY_NAME \
  --query loginServer \
  --output tsv)

echo "  Login server: $LOGIN_SERVER"

# Get admin credentials
echo ""
echo "Admin credentials:"
az acr credential show \
  --name $REGISTRY_NAME \
  --query "{username: username, password: passwords[0].value}" \
  --output table

echo ""
echo "✅ Container Registry created successfully."
echo "   Login server: $LOGIN_SERVER"
echo "   View in portal: https://portal.azure.com/#resource/subscriptions/$SUBSCRIPTION_ID/resourceGroups/$RESOURCE_GROUP/providers/Microsoft.ContainerRegistry/registries/$REGISTRY_NAME"
````

## Cell 3: Build and Push Image to ACR

**Goal:** Build Docker image and push to Azure Container Registry.

**What we're doing:**
1. Login to ACR using Azure CLI
2. Tag local image with ACR login server
3. Push image to ACR
4. Verify image in registry

**Alternative approach:** Use `az acr build` to build directly in Azure (no local Docker required).

````python
%%bash -s "$AZURE_CONTAINER_REGISTRY" "$IMAGE_NAME" "$IMAGE_TAG"

REGISTRY_NAME=$1
IMAGE_NAME=$2
IMAGE_TAG=$3

# Get login server
LOGIN_SERVER=$(az acr show \
  --name $REGISTRY_NAME \
  --query loginServer \
  --output tsv)

echo "Login server: $LOGIN_SERVER"

# Login to ACR (uses Azure CLI credentials)
echo ""
echo "Logging in to ACR..."
az acr login --name $REGISTRY_NAME

# Tag local image for ACR
echo ""
echo "Tagging image: $IMAGE_NAME:$IMAGE_TAG -> $LOGIN_SERVER/$IMAGE_NAME:$IMAGE_TAG"
docker tag $IMAGE_NAME:$IMAGE_TAG $LOGIN_SERVER/$IMAGE_NAME:$IMAGE_TAG
docker tag $IMAGE_NAME:$IMAGE_TAG $LOGIN_SERVER/$IMAGE_NAME:latest

# Push images to ACR
echo ""
echo "Pushing images to ACR..."
docker push $LOGIN_SERVER/$IMAGE_NAME:$IMAGE_TAG
docker push $LOGIN_SERVER/$IMAGE_NAME:latest

# List images in registry
echo ""
echo "Images in ACR:"
az acr repository list \
  --name $REGISTRY_NAME \
  --output table

echo ""
echo "Image tags:"
az acr repository show-tags \
  --name $REGISTRY_NAME \
  --repository $IMAGE_NAME \
  --output table

echo ""
echo "✅ Image pushed successfully."
echo "   Full image path: $LOGIN_SERVER/$IMAGE_NAME:$IMAGE_TAG"
````

## Cell 4: Deploy to Azure Container Instances (ACI)

**Goal:** Run Flask container on Azure Container Instances.

**What we're doing:**
1. Get ACR credentials
2. Create container instance with ACR image
3. Expose port 5000 with public IP
4. Test the deployed app

**ACI use cases:**
- Quick testing/demos (no infrastructure setup)
- Batch jobs (run once and terminate)
- CI/CD ephemeral environments
- **Not for production web apps** (use App Service or AKS for high availability)

````python
%%bash -s "$AZURE_RESOURCE_GROUP" "$AZURE_CONTAINER_REGISTRY" "$IMAGE_NAME" "$IMAGE_TAG" "$AZURE_LOCATION"

RESOURCE_GROUP=$1
REGISTRY_NAME=$2
IMAGE_NAME=$3
IMAGE_TAG=$4
LOCATION=$5

CONTAINER_NAME="flask-aci-demo"

# Get ACR credentials
LOGIN_SERVER=$(az acr show --name $REGISTRY_NAME --query loginServer -o tsv)
USERNAME=$(az acr credential show --name $REGISTRY_NAME --query username -o tsv)
PASSWORD=$(az acr credential show --name $REGISTRY_NAME --query "passwords[0].value" -o tsv)

echo "Creating container instance: $CONTAINER_NAME"
echo "  Image: $LOGIN_SERVER/$IMAGE_NAME:$IMAGE_TAG"
echo "  CPU: 1 core, Memory: 1.5 GB"

# Create container instance
az container create \
  --resource-group $RESOURCE_GROUP \
  --name $CONTAINER_NAME \
  --image $LOGIN_SERVER/$IMAGE_NAME:$IMAGE_TAG \
  --registry-login-server $LOGIN_SERVER \
  --registry-username $USERNAME \
  --registry-password $PASSWORD \
  --cpu 1 \
  --memory 1.5 \
  --ports 5000 \
  --dns-name-label flask-demo-$(openssl rand -hex 4) \
  --restart-policy OnFailure

# Get container details
echo ""
echo "Container details:"
az container show \
  --resource-group $RESOURCE_GROUP \
  --name $CONTAINER_NAME \
  --query "{FQDN: ipAddress.fqdn, IP: ipAddress.ip, State: instanceView.state}" \
  --output table

# Get FQDN and test endpoint
FQDN=$(az container show \
  --resource-group $RESOURCE_GROUP \
  --name $CONTAINER_NAME \
  --query ipAddress.fqdn \
  --output tsv)

echo ""
echo "Waiting 10 seconds for container to start..."
sleep 10

echo ""
echo "Testing deployed app:"
curl -s http://$FQDN:5000/health | python -m json.tool

echo ""
echo "✅ Container deployed successfully."
echo "   URL: http://$FQDN:5000"
echo "   Test: curl http://$FQDN:5000/"
````

## Cell 5: Monitor Container with Azure Container Insights

**Goal:** View logs and metrics for deployed container.

**What we're doing:**
1. Stream container logs (stdout/stderr)
2. Check container metrics (CPU, memory)
3. View events (restarts, errors)
4. Cleanup resources

**Production monitoring:** For production workloads, enable **Container Insights** (Azure Monitor integration) for advanced metrics, alerting, and log aggregation.

````python
%%bash -s "$AZURE_RESOURCE_GROUP"

RESOURCE_GROUP=$1
CONTAINER_NAME="flask-aci-demo"

echo "Container logs (last 20 lines):"
az container logs \
  --resource-group $RESOURCE_GROUP \
  --name $CONTAINER_NAME \
  --tail 20

echo ""
echo "Container state and events:"
az container show \
  --resource-group $RESOURCE_GROUP \
  --name $CONTAINER_NAME \
  --query "instanceView.{State: state, Restart: restartCount, Events: events[].{Type: type, Message: message}}" \
  --output yaml

echo ""
echo "---"
echo ""
echo "🧹 Cleanup (delete all resources to avoid charges):"
echo ""
echo "Delete container instance:"
echo "  az container delete --resource-group $RESOURCE_GROUP --name $CONTAINER_NAME --yes"
echo ""
echo "Delete entire resource group (ACR + ACI):"
echo "  az group delete --name $RESOURCE_GROUP --yes --no-wait"
echo ""
echo "⚠️  Resource group deletion is PERMANENT and cannot be undone."
echo "   Uncomment and run the following to delete:"
echo ""
# Uncomment to execute cleanup:
# az group delete --name $RESOURCE_GROUP --yes --no-wait
# echo "✅ Cleanup initiated. Resources will be deleted in ~5 minutes."
````
