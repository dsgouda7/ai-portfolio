```python
import torch

def check_gpu_availability():
    """Check if GPU is available and warn if not."""
    if not torch.cuda.is_available():
        print("⚠️  WARNING: No GPU detected!")
        print("   This notebook requires a CUDA-capable GPU.")
        print("   CFG scaling computationally expensive (8GB+ VRAM recommended)")
        print("   Options:")
        print("   1. Run on Google Colab (free T4 GPU)")
        print("   2. Use Kaggle Notebooks (free P100 GPU)")
        print("   3. Rent cloud GPU (Lambda Labs, RunPod)")
        print("   4. Continue on CPU (10-50x slower)")
        print()
        response = input("Continue anyway? (yes/no): ")
        if response.lower() != 'yes':
            raise RuntimeError("GPU required. Exiting.")
    else:
        gpu_name = torch.cuda.get_device_name(0)
        vram_gb = torch.cuda.get_device_properties(0).total_memory / 1e9
        print(f"✅ GPU detected: {gpu_name} ({vram_gb:.1f} GB VRAM)")

check_gpu_availability()
```
