# CPU Text-to-Speech Quick Win

This notebook shows a minimal text-to-speech demo using a pretrained model that runs on a stock CPU (no GPU required).

Model used: `facebook/mms-tts-eng` (Hugging Face Transformers).

```python
%pip install -q transformers torch scipy
```

```python
import torch
from transformers import AutoTokenizer, VitsModel
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio, display

MODEL_ID = "facebook/mms-tts-eng"
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = VitsModel.from_pretrained(MODEL_ID).to("cpu")
sample_rate = model.config.sampling_rate

print(f"Loaded {MODEL_ID} on CPU at {sample_rate} Hz")
```

```python
def speak(text: str, out_file: str = "quickwin_tts.wav"):
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        waveform = model(**inputs).waveform.squeeze().cpu().numpy()

    write_wav(out_file, rate=sample_rate, data=waveform)
    print(f"Saved {out_file}")
    display(Audio(waveform, rate=sample_rate))

speak("Hello from PixelSmith. This speech was generated on a CPU-only notebook.")
```

```python
# Try your own prompt in one line:
speak("Replace this text with your own message for a quick audio demo.", "custom_tts.wav")
```
